```python
import matplotlib.pyplot as plt
import math

# Define the Lissajous function
def lisj(t, A=1, T=100, phi=0.0):
    """
    Lissajous function: sin(2Ï€ * t / T + phi)
    t : time
    T : period (controls frequency)
    phi : phase shift
    """
    phase = 2*math.pi/T*t
    phase += phi
    x = A * math.sin(phase)
    return x

# Define helper plotting function with grid
def plot_with_layout(x, y, ax=None):
    if ax is None:
        ax = plt.gca()
    ax.grid()
    ax.set_xticks([-1, -0.5, 0, 0.5, 1])
    ax.set_yticks([-1, -0.5, 0, 0.5, 1])
    ax.set_aspect('equal')
    ax.plot(x, y)

    # Label top row with nx
    if ny == 1:
        ax.set_title(f"nx={nx}", fontsize=8)

    # Label left column with ny
    if nx == 1:
        ax.set_ylabel(f"ny={ny}", fontsize=8, rotation=0, labelpad=20)

# Main parameters
T = 100
t = [*range(0, 50*T+1)]
px = 0.0 * math.pi
py = 0.0 * math.pi

# Create figure and axes
fig, axes = plt.subplots(6, 6, figsize=(8, 8))

for ny in range(1, 7):
    for nx in range(1, 7):
        ax = axes[ny-1, nx-1]
        x = [lisj(j, T=nx*T, phi=px) for j in t]
        y = [lisj(j, T=ny*T, phi=py) for j in t]
        plot_with_layout(x, y, ax=ax)
        ax.tick_params(axis='x', labelbottom=False)
        ax.tick_params(axis='y', labelleft=False)
plt.tight_layout()
plt.show()
```
